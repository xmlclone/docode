# yield表达式

```python
def f1():
    value = 0
    while True:
        value = yield value + 1
        if value is None:
            value = 0
        print(f"You send {value=}")

f = f1()
f
```

```python
# 会进行第一次循环
next(f)
```

```python
next(f)
```

```python
f.send(2)
```

```python
from typing import Generator


def f1() -> Generator[int, None, None]:
    print("f1 start.")
    yield 1
    print("f2 end.")

def f2():
    return f1()

# f2()返回的是f1()，是个generator，未通过next
f2()
```
